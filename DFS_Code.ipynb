```python
graph_1 = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}
```

```python
graph_2 = {
    'u': ['v', 'x'],
    'v': ['y'],
    'w': ['y', 'z'],
    'x': ['v'],
    'y': ['x'],
    'z': ['z']  
}
```

```python
time = 0
color ={}
parent = {} 
discovery = {}
finish = {}

def dfs(G):
    for i in G:
        color[i] = "White"
        parent[i] = None

    for i in G:
        if color[i] == "White":
            dfs_visit(G,i)

def dfs_visit(G,i):
    global time
    color[i] = "Grey"
    time +=1
    discovery[i] = time

    for v in G[i]:
        if color[v] == "White":
            parent[v] = i
            dfs_visit(G,v)
    
    color[i] = "Black"
    time +=1
    finish[i] = time

dfs(graph_1)

# --- Print Results ---
print("Discovery Times:", discovery)
print("Finish Times:", finish)
print("Parents:", parent)
```
