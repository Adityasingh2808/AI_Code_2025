```python
graph = {
    'A': ['B', 'C'],
    'B': ['D', 'E'],
    'C': ['F'],
    'D': [],
    'E': ['F'],
    'F': []
}
```

```python
from collections import deque
def bfs(graph, start):
    color ={i: "white" for i in graph}
    distance = {i: float("inf") for i in graph}
    parent = {i : None for i in graph}

    color[start] = "gray"
    distance[start] = 0
    parent[start] = None

    Q = deque()
    Q.append(start)

    while Q:
        u = Q.popleft()
        for i in graph[u]:
            if color[i] == "white":
                color[i] = "gray"
                distance[i] = distance[u]+1
                parent[i] = u 
                Q.append(i)
        color[u] = "black"
    print(color)
    return distance 
bfs(graph,"A")
```
