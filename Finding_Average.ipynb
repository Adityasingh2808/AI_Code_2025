```python
sum=0
for j in range (10):
    result=0
    import random
    i=0
    while (i<1000):
        a=random.random()
        if a<=0.5:
            a=1
        else:
            a=0
        b=random.random()
        if b<=0.4:
            b=1
        else:
            b=0
        c=a*b
        result+=c
        i+=1
    print("run ",j ,":- no of days when both a and b have attended class ",result)
    sum+=result
print("Average: ", sum/10000)
```
